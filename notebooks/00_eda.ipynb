```python
def check_and_drop_duplicates(df, name="dataset"):
    n_dups = df.duplicated().sum()
    print(f"\n{name} : {n_dups} doublons détectés")
    if n_dups > 0:
        df_clean = df.drop_duplicates().reset_index(drop=True)
        print(f"→ Après suppression : {df_clean.shape[0]} lignes")
        return df_clean
    else:
        print("✅ Aucun doublon trouvé")
        return df


X_train = check_and_drop_duplicates(X_train, "X_train")
y_train = check_and_drop_duplicates(y_train, "y_train")
X_test  = check_and_drop_duplicates(X_test, "X_test")

```

```python
features = [c for c in X_train.columns if c != "ID"]
merged = X_train.copy()
merged["target_tuple"] = y_train.apply(tuple, axis=1)

dup_targets = merged.groupby(features)["target_tuple"].nunique()
print("Nombre de combinaisons features avec plusieurs targets :", (dup_targets > 1).sum())
```

```python
import numpy as np

def detect_outliers_iqr(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    return ((df[col] < lower) | (df[col] > upper)).sum()

for col in [c for c in X_train.columns if c not in ["ID"]]:
    outliers = detect_outliers_iqr(X_train, col)
    if outliers > 0:
        print(f"{col} → {outliers} outliers détectés")
```

```python
def clip_outliers(df, cols, lower_q=0.01, upper_q=0.99):
    for col in cols:
        low = df[col].quantile(lower_q)
        high = df[col].quantile(upper_q)
        df[col] = df[col].clip(low, high)
    return df

X_train_clipped = clip_outliers(X_train.copy(), [c for c in X_train.columns if c != "ID"])
X_test_clipped  = clip_outliers(X_test.copy(), [c for c in X_test.columns if c != "ID"])
```
